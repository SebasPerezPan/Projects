```python
from connector import *
from main import *
import pandas as pd
from bokeh.io import output_notebook
output_notebook()
```

```python
competition_list = sorted(get_competitions(connection))
for competition in competition_list:
    print(f"{competition[0]}. {competition[1].replace('_', ' ').title()}")
competition_id = int(input("Selecciona una competición:"))
competition_name = competition_list[competition_id-1][1]
print(f"Has seleccionado {competition_name.replace('_', ' ').title()}")
```

```python
season_list = list(get_seasons(connection, competition_id))
contador = 0
for season in season_list:
    contador += 1
    print(f"{contador}. {season[2].replace('_','/')}")
if len(season_list) > 1:
    season_option = int(input("Selecciona una temporada:"))
else:
    season_option = 1
season_id = season_list[season_option - 1][0]
season_name = season_list[season_option-1][2]
print(f"Has seleccionado {season_name.replace('_', '/').title()}")
```

```python
selected_season = Season(season_id, season_name, competition_id, competition_name)
selected_season.load_teams_and_players(connection)
selected_season.load_matches_and_assign_to_teams(connection)
```

```python
selected_season.display_top_players()
```

```python
print("\nEquipos disponibles en la temporada:")
for idx, team in enumerate(selected_season.teams):
    print(f"{idx + 1}. {team.team_name}")

# Seleccionar equipos por índice
try:
    selected_main_team_idx = int(input("Selecciona el equipo principal (por número): ")) - 1
    selected_secondary_team_idx = int(input("Selecciona el equipo secundario (por número): ")) - 1

    # Obtener los objetos de los equipos seleccionados
    selected_main_team = selected_season.teams[selected_main_team_idx]
    selected_secondary_team = selected_season.teams[selected_secondary_team_idx]

    print(f"\nEquipo principal seleccionado: {selected_main_team.team_name}")
    print(f"Equipo secundario seleccionado: {selected_secondary_team.team_name}")

except IndexError:
    print("Seleccionaste un número fuera del rango. Por favor, inténtalo de nuevo.")
except ValueError:
    print("Entrada inválida. Por favor, ingresa un número.")
```

```python
selected_main_team.display_team_summary()
selected_secondary_team.display_team_summary()
```

```python
selected_main_team.display_top_players()
selected_secondary_team.display_top_players()
```

```python
print(f"\nJugadores del equipo {selected_main_team.team_name}:")
for idx, player in enumerate(selected_main_team.players):
    print(f"{idx + 1}. {player.player_name} - Posición: {player.position}")

# Seleccionar un jugador por índice
try:
    selected_player_idx = int(input("Selecciona un jugador (por número): ")) - 1
    selected_player = selected_main_team.players[selected_player_idx]
    
    print(f"\nJugador seleccionado: {selected_player.player_name}")
    
    # Mostrar estadísticas detalladas del jugador
    selected_player.display_detailed_stats(connection)
    
except IndexError:
    print("Seleccionaste un número fuera del rango. Por favor, inténtalo de nuevo.")
except ValueError:
    print("Entrada inválida. Por favor, ingresa un número.")
```
